# Iris Prediction Example
Load the saved model and predict a flower species.

```

import joblib

# Load trained model
model = joblib.load("../model/knn_iris_model.sav")
```

```

# Example input:
# [sepal_length, sepal_width, petal_length, petal_width]

test_data = [[6.8, 3.0, 1.7, 0.5]]

prediction = model.predict(test_data)

print("Predicted flower species:", prediction[0])
```
